```python
df = pd.concat([pd.read_excel(f, encoding='latin1') for f in glob.glob('*.xls')])
```

```python
# DF TO EXCEL
from pandas import ExcelWriter

writer = ExcelWriter('Latrocinio.xlsx')
df.to_excel(writer,'Sheet1')
writer.save()
```

```python
# DF TO Pickle  (in order to keep df format)

import pickle

outfile = open('Latrocinio.pickle','wb')
pickle.dump(df, outfile)
outfile.close()
```
